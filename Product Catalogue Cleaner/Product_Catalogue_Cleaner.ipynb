```python
#Raw Data
raw_Products = [
    {"product_id": "101", "name": "Milk", "price": "120", "category": "dairy"},
    {"product_id": 102, "name": "Cheese", "price": 450.0, "category": "dairy"},
    {"product_id": "103", "name": "Yogurt", "price": None, "category": "dairy"},
    {"product_id": 104, "name": "Apple", "price": "220.5", "category": "fruits"},
    {"product_id": "105", "name": "Banana", "price": 180, "category": "fruits"},
    {"product_id": 106, "name": "", "price": "75", "category": "vegetables"},
    {"product_id": None, "name": "Carrot", "price": "90", "category": "vegetables"},
    {"product_id": "108", "name": "Spinach", "price": "N/A", "category": "vegetables"},
    {"product_id": 109, "name": "Mango", "price": 300, "category": "fruits"},
    {"product_id": "110", "name": "Butter", "price": "-50", "category": "dairy"},
    {"product_id": 104, "name": "Apple", "price": "220.5", "category": "fruits"},
    {"product_id": "112", "name": "Orange", "price": 250.75, "category": "fruits"},
    {"product_id": 113, "name": "Tomato", "price": "85.0", "category": "vegetables"},
    {"product_id": "114", "name": None, "price": "60", "category": "vegetables"},
    {"product_id": 115, "name": "Cream", "price": 200, "category": "dairy"}
]
```

```python
#Cleaning

from types import NoneType

sum = 0

for i in range(len(raw_Products)):
  if not isinstance(raw_Products[i]["price"], float):
    if isinstance(raw_Products[i]["price"], str):
      raw_Products[i]["price"] = raw_Products[i]["price"].replace("$", "")
    try:
      raw_Products[i]["price"] = float(raw_Products[i]["price"])
    except:
      raw_Products[i]["price"] = 0.00

    if raw_Products[i]["price"] < 0:
      raw_Products[i]["price"] = 0.00

  if not isinstance(raw_Products[i]["product_id"], int):
    if isinstance(raw_Products[i]["product_id"], NoneType):
      raw_Products[i]["product_id"] = 0
    raw_Products[i]["product_id"] = int(raw_Products[i]["product_id"])

  if isinstance(raw_Products[i]["name"], NoneType):
      raw_Products[i]["name"] = ""

  sum += raw_Products[i]["price"]
```

```python
#Dictionary for summary
organizedData = { "Dairy": [],
                  "Fruits": [],
                  "Vegetables": []
}

for j in range(len(raw_Products)):

  if raw_Products[j]["category"] == "dairy":
    organizedData["Dairy"].append(raw_Products[j])

  elif raw_Products[j]["category"] == "fruits":
    organizedData["Fruits"].append(raw_Products[j])

  else:
    organizedData["Vegetables"].append(raw_Products[j])
```

```python
#Finding Duplicate

temp = organizedData.keys()
cat = list(temp)
ids = list()

for i in range(len(organizedData.keys())):
  for j in range(len(organizedData[cat[i]])):
    ids.append(organizedData[cat[i]][j]["product_id"])

ids.sort()

print("Duplicate: ")
for k in range(len(ids)):
  if ids[k] == ids[k-1]:
    print(ids[k])
    dup = ids[k]

for i in range(len(organizedData.keys())):
  for j in range(len(organizedData[cat[i]])):
    if organizedData[cat[i]][j]["product_id"] == dup:
      organizedData[cat[i]].pop(j)
      break
```

```python
#Printing Summary

print("< Overall Summary >")
print("Total Categories: ",len(organizedData.keys()))
print("Total Unique Products: ",len(organizedData["Dairy"] + organizedData["Fruits"] + organizedData["Vegetables"]))
print("Total Amount: ",sum)
print("Total Dairy Products:",len(organizedData["Dairy"]))
print("Total Fruits Products:",len(organizedData["Fruits"]))
print("Total Vegetables Products:",len(organizedData["Vegetables"]))
```
