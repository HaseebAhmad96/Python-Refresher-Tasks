```python
import csv
import os

results = []

#Ranges foe marks
mark_ranges = {
    2: (0,10),
    3: (0,10),
    4: (0,30),
    5: (0,10),
    6: (0,40)
}

class MarkValidationError(Exception):
    pass

folder_path = '/content/marks'
files = os.listdir(folder_path)
```

```python
try:
  if os.path.exists(folder_path):
    print(f"Folder '{folder_path}' found.")

    for file in files:
      if file.endswith('.csv'):
        file_path = os.path.join(folder_path, file)
        print(f"{file} found")

        try:
          with open(file_path, 'r') as csvf:
            csv_reading = csv.reader(csvf)
            head = next(csv_reading)
            data = list(csv_reading)

            for row in data:
              current_status = "Unknown"

              for i in range(len(row)):
                if i > 1 and i < 7:
                  if not isinstance(row[i],(float,int)):
                    if isinstance(row[i],str):
                      if row[i] == "N/A" or row[i] == "":
                        row[i] = 0.0
                      else:
                        try:
                          row[i] = float(row[i])
                        except ValueError:
                          row[i] = 0.0

                  if row[i] is None:
                    row[i] = 0.0

                  if row[i] < 0:
                    raise MarkValidationError("Marks cannot be negative")

                  if i == 2 or i == 3 or i == 5:
                    if row[i] > 10:
                      raise MarkValidationError("Quiz/Assignment marks out of range")

                  if i == 4:
                    if row[i] > 30:
                      raise MarkValidationError("Mids marks out of range")

                  if i == 6:
                    if row[i] > 40:
                      raise MarkValidationError("Final marks out of range")


                elif i == 7:
                  if not isinstance(row[i],(float,int)):
                    if isinstance(row[i],str):
                      if row[i] == "N/A" or row[i] == "":
                        row[i] = 0.0
                      else:
                        try:
                          row[i] = float(row[i])
                        except ValueError:
                          row[i] = 0.0

                  if row[i] is None:
                    row[i] = 0.0

                  if row[i] > 100:
                    raise MarkValidationError("Total marks out of range")


              name = row[0]
              roll_no = row[1]
              total_marks = row[7]


              if total_marks >= 50:
                current_status = "Pass"
              else:
                current_status = "Fail"


              results.append([name, roll_no, total_marks, current_status])

        except FileNotFoundError:
          print("File not found")

        except MarkValidationError as e:
          print("Mark validation error:", e)
        except Exception as e:
          print(f"An error occurred while processing {file}: {e}")

      else:
        print("File is not CSV")

except Exception as e:
  print(f"An unexpected error occurred: {e}")
```

```python
with open("Results.csv", "w", newline="") as f:
  writer = csv.writer(f)
  writer.writerow(["Name","Roll No","Total Marks","Status"])

  for r in results:
    writer.writerow(r)
```
