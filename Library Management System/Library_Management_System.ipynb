```python
class Book:

    def __init__(self, title, author, isbn):

        self.title = title
        self.author = author
        self.isbn = isbn
        self.available = True


    #Readable output
    def __str__(self):

        if self.available:
            status = "Available"
        else:
            status = "Borrowed"

        return f"{self.title} by {self.author} ({status})"


    #Compare books by isbn
    def __eq__(self, other):

        return self.isbn == other.isbn
```

```python
class Member:

    def __init__(self, name):

        self.name = name
        self.borrowed_books = []


    def borrow(self, book):

        if not book.available:
            print("Book not available")
            return

        if len(self.borrowed_books) >= self.borrow_limit():
            print("Borrow limit reached")
            return

        book.available = False
        self.borrowed_books.append(book)

        print(self.name, "borrowed", book.title)


    def return_book(self, book):

        if book in self.borrowed_books:

            self.borrowed_books.remove(book)
            book.available = True

            print(self.name, "returned", book.title)

        else:
            print("This book was not borrowed")


    def borrow_limit(self):

        return 2


    def calculate_fine(self, days):

        return days * 1


    def __str__(self):

        return f"{self.name} | Books borrowed: {len(self.borrowed_books)}"
```

```python
class PremiumMember(Member):

    def borrow_limit(self):

        return 5


    def calculate_fine(self, days):

        return days * 0.5


class GoldMember(PremiumMember):

    def borrow_limit(self):

        return 10


    def calculate_fine(self, days):

        return 0
```

```python
class Library:

    def __init__(self, name):

        self.name = name
        self.books = []


    def add_book(self, book):

        self.books.append(book)


    def show_books(self):

        print("\nAvailable books:")

        for book in self.books:

            if book.available:
                print(book)


    def __str__(self):

        return f"Library: {self.name} | Total books: {len(self.books)}"
```

```python
b1 = Book("Clean Code", "Robert Martin", "ISBN1")
b2 = Book("Python Crash Course", "Eric Matthes", "ISBN2")
b3 = Book("Fluent Python", "Luciano Ramalho", "ISBN3")

library = Library("City Library")

library.add_book(b1)
library.add_book(b2)
library.add_book(b3)

print(library)
library.show_books()
```

```python
# Members
haseeb = Member("Haseeb")
arslan = PremiumMember("Arslan")
umair = GoldMember("Umair")

print("\nMembers")
print(haseeb)
print(arslan)
print(umair)


print("\nBorrowing")
haseeb.borrow(b1)
arslan.borrow(b2)
umair.borrow(b3)   # limit reached
```

```python
arslan.borrow(b3)

library.show_books()


print("\nReturning book")
haseeb.return_book(b1)

library.show_books()


print("\nFine examples")
print("Haseeb fine for 5 days:", haseeb.calculate_fine(5))
print("Arslan fine for 5 days:", arslan.calculate_fine(5))
print("Umair fine for 5 days:", umair.calculate_fine(5))
```
