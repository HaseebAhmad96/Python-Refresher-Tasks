```python
employees = [
    {"name": "Ahmad","score": 88,"years": 5},
    {"name": "Omar","score": 72,"years": 8},
    {"name": "Ali","score": 95,"years": 3},
    {"name": "Khalid","score": 88,"years": 7},
    {"name": "Abu Bakr","score": 61,"years": 2},
    {"name": "Arslan","score": 95,"years": 3},
    {"name": "Umair","score": 77,"years": 6},
    {"name": "Fatmia","score": 72,"years": 8},
    {"name": "Rabia","score": 84,"years": 4},
    {"name": "Sidra","score": 61,"years": 9},
]
```

```python
def ranking_key(emp):
    return (-emp["score"], -emp["years"], emp["name"])


ranked = sorted(employees, key=ranking_key)


print("Employee Rankings")
print("Rank Name Score Years")

for i, emp in enumerate(ranked, start=1):
    print(i, emp["name"], emp["score"], emp["years"])
```

```python
def name_key(emp):
    return emp["name"]

name_sorted = sorted(employees, key=name_key)

name_list = []
for emp in name_sorted:
    name_list.append(emp["name"])
```

```python
def binary_search(names, target):

    low = 0
    high = len(names) - 1

    while low <= high:

        mid = (low + high) // 2

        if names[mid] == target:
            return mid

        elif target < names[mid]:
            high = mid - 1

        else:
            low = mid + 1

    return -1


def find_employee(name):

    index = binary_search(name_list, name)

    if index == -1:
        print("\nEmployee not found")
        return
    emp = name_sorted[index]

    print("\nEmployee Found")
    print("Name :", emp["name"])
    print("Score:", emp["score"])
    print("Years:", emp["years"])
```

```python
find_employee("Arslan")
find_employee("Umair")
find_employee("Haseeb")
```
