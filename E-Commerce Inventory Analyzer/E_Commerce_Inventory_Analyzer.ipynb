```python
products = [
    {"name": "Apple","category": "Fruits","price": 1.20},
    {"name": "Banana","category": "fruits","price": 0.50},
    {"name": "Laptop","category": "Electronics", "price": 999.99},
    {"name": "Phone","category": "Electronics", "price": 799.00},
    {"name": "Apple",  "category": "Fruits",  "price": 1.20},
    {"name": "Headphones", "category": "Electronics", "price": 199.99},
    {"name": "Mango",  "category": "Fruits",  "price": None},
    {"name": "Shirt",  "category": "Clothing",  "price": 29.99},
    {"name": "Jeans",  "category": "clothing",  "price": 59.99},
    {"name": "Monitor",  "category": "Electronics", "price": 349.99},
    {"name": "Jacket",  "category": "Clothing",  "price": 89.99},
    {"name": "Grapes", "category": "Fruits", "price": 2.50},
    {"name": "Tablet", "category": "Electronics", "price": 499.00},
    {"name": "Banana",  "category": "Fruits",  "price": 0.50},
]
```

```python
cleaned = [  {**p, "category": p["category"].title()}  for p in products if p["price"] is not None   ]

seen = set()
unique = []

for p in cleaned:
    key = (p["name"], p["category"])

    if key not in seen:
        seen.add(key)
        unique.append(p)

print(f"Original: {len(products)} items After cleaning: {len(unique)} items\n")

categories = {cat: [] for cat in {p["category"] for p in unique}}
```

```python
def price_func(products):
  return p["price"]

for p in unique:
    categories[p["category"]].append(p)

top5 = { cat: sorted(items, key=price_func, reverse=True)[:5] for cat, items in categories.items() }
```

```python
summary = {
    cat: {
        "total_products": len(items),
        "avg_price":      round(sum(p["price"] for p in items) / len(items), 2),
        "top_5":          [f"{p['name']} (${p['price']})" for p in top5[cat]],
    }
    for cat, items in categories.items()
}
```

```python
print("Summary:")

for cat, info in summary.items():
    print(f"\n {cat}")
    print(f"Total products: {info['total_products']}")
    print(f"Average price: ${info['avg_price']}")
    print(f"Top 5 priciest:")
    for item in info["top_5"]:
        print(f"{item}")
```
