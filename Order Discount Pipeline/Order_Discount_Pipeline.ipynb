```python
from functools import reduce
orders = [
    {"id":1,"customer":"Alice","category":"Electronics","amount":1200.00},
    {"id":2,"customer":"Bob","category":"Clothing","amount":85.00},
    {"id":3,"customer":"Charlie","category":"Food","amount":40.00},
    {"id":4,"customer":"Diana","category":"Electronics","amount":300.00},
    {"id":5,"customer":"Eve","category":"Clothing","amount":25.00},
    {"id":6,"customer":"Frank","category":"Food","amount":15.00},
    {"id":7,"customer":"Grace","category":"Electronics","amount":650.00},
    {"id":8,"customer":"Hank","category":"Clothing","amount":200.00},
]
```

```python
DISCOUNTS = {
    "Electronics":0.10,
    "Clothing":0.15,
    "Food":0.05
}

MIN_ORDER = 50.0

def apply_discount(order):

    category = order["category"]
    amount = order["amount"]

    discount = DISCOUNTS.get(category,0)

    final_amount = amount * (1 - discount)

    new_order = {
        "id":order["id"],
        "customer":order["customer"],
        "category":category,
        "original_amount":amount,
        "discount":discount,
        "final_amount":round(final_amount,2)
    }

    return new_order
```

```python
discounted_orders = list(map(apply_discount, orders))

def is_valid(order):

    if order["final_amount"] >= MIN_ORDER:
        return True
    else:
        return False


valid_orders = list(filter(is_valid, discounted_orders))

def sort_by_amount(order):

    return order["final_amount"]


sorted_orders = sorted(valid_orders, key=sort_by_amount, reverse=True)
```

```python
def add_revenue(total, order):

    return total + order["final_amount"]


total_revenue = reduce(add_revenue, sorted_orders, 0)

def create_summary(order):

    summary = {
        "id":order["id"],
        "customer":order["customer"],
        "category":order["category"],
        "original":order["original_amount"],
        "discount_percent": int(order["discount"] * 100),
        "final":order["final_amount"]
    }

    return summary
```

```python
summary = list(map(create_summary, sorted_orders))

print("Processed Orders\n")

for i, order in enumerate(summary,1):

    print("Order",i)
    print("Customer:",order["customer"])
    print("Category:",order["category"])
    print("Original Amount:",order["original"])
    print("Discount:",order["discount_percent"],"%")
    print("Final Amount:",order["final"])
    print()


print("Orders dropped:", len(orders) - len(valid_orders))
print("Orders processed:", len(valid_orders))
print("Total revenue:", total_revenue)
```
